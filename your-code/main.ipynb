#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.

```python
# MEDIAN STEPS BY MONTH_ENCODED
median_steps_by_month_encoded = fitbit.groupby('Months_encoded')['Steps'].median().reset_index()
print(median_steps_by_month_encoded)                                       
```

```python
# MEDIAN STEPS BY WEEK_OR_WEEKEND
median_steps_work_or_weekend = fitbit.groupby('Work_or_Weekend')['Steps'].median().reset_index()
print(median_steps_work_or_weekend)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4)) 

# median steps by month
axes[0].bar(median_steps_by_month_encoded['Months_encoded'], median_steps_by_month_encoded['Steps'], color='skyblue')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Median Steps')
axes[0].set_title('Median Steps by Month')
axes[0].set_xticks(range(1, 13))

#median steps by work and weekend
axes[1].bar(median_steps_work_or_weekend['Work_or_Weekend'], median_steps_work_or_weekend['Steps'], color='lightgreen')
axes[1].set_xlabel('Weekday (1) or Weekend (0)')
axes[1].set_ylabel('Median Steps')
axes[1].set_title('Median Steps by Weekday/Weekend')
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(['Weekend', 'Weekday'])


```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
feature_combinations = [
    ('Minutes Lightly Active', 'Steps'),
    ('Minutes Very Active', 'Steps'),
    ('Minutes Sedentary', 'Steps')
]

# looping
for i, (x_feature, y_feature) in enumerate(feature_combinations, start=1):
    plt.figure(figsize=(8, 6))
    x_data = fitbit[x_feature]
    y_data = fitbit[y_feature]
    
    # scatter
    plt.scatter(x_data, y_data, alpha=0.5)
    
    # labels
    plt.title(f'Scatter Plot {i}: {x_feature} vs {y_feature}')
    plt.xlabel(x_feature)
    plt.ylabel(y_feature)
    plt.grid(True)
    plt.savefig(f'scatter_plot_{i}.png')
    plt.show()
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
from sklearn import linear_model
x = titanic[['Age']]  # independent variable(s)
y = titanic['Fare']   # dependent variable
model = linear_model.LinearRegression()
result = model.fit(x, y)
```

```python
regression_line = result.intercept_ + titanic.Age * result.coef_
plt.plot(titanic.Age, regression_line, c='orange')
plt.scatter(titanic.Age, titanic.Fare)
plt.show()
```

```python
# Method 2 - seaborn
x = titanic['Age']
y = titanic['Fare']
sns.regplot(x=x, y=y, scatter_kws={'color': 'blue'}, line_kws={'color': 'green'})
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Regression Plot of Fare vs. Age')
plt.show()
```
