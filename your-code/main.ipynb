#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 6)) 

# x and y
sns.lineplot(data=data, x='x', y='y', ax=axes[0], color='blue', label='y')
axes[0].set_xlabel('X-axis')
axes[0].set_ylabel('Y-axis')
# x and z
sns.lineplot(data=data, x='x', y='z', ax=axes[1], color='red', label='z')
axes[1].set_xlabel('X-axis')
axes[1].set_ylabel('Z-axis')
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 6)) 

# x and y
sns.lineplot(data=data, x='x', y='y', ax=axes[0], color='blue', label='y')
axes[0].set_xlabel('X-axis')
axes[0].set_ylabel('Y-axis')
axes[0].grid()
axes[0].legend()

# x and z
sns.lineplot(data=data, x='x', y='z', ax=axes[1], color='red', label='z')
axes[1].set_xlabel('X-axis')
axes[1].set_ylabel('Z-axis')
axes[1].grid()
axes[1].legend()

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
median_steps_by_month_encoded = fitbit.groupby('Months_encoded')['Steps'].median().reset_index()
print(median_steps_by_month_encoded)                                       
```

```python
# MEDIAN STEPS BY WEEK_OR_WEEKEND
median_steps_work_or_weekend = fitbit.groupby('Work_or_Weekend')['Steps'].median().reset_index()
print(median_steps_work_or_weekend)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4)) 

# median steps by month
axes[0].bar(median_steps_by_month_encoded['Months_encoded'], median_steps_by_month_encoded['Steps'], color='skyblue')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Median Steps')
axes[0].set_title('Median Steps by Month')
axes[0].set_xticks(range(1, 13))

#median steps by work and weekend
axes[1].bar(median_steps_work_or_weekend['Work_or_Weekend'], median_steps_work_or_weekend['Steps'], color='lightgreen')
axes[1].set_xlabel('Weekday (1) or Weekend (0)')
axes[1].set_ylabel('Median Steps')
axes[1].set_title('Median Steps by Weekday/Weekend')
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(['Weekend', 'Weekday'])

#scatter
plt.sca(axes[0])
plt.sca(axes[1])
```

```python
plt.scatter(y=median_steps_by_month_encoded, x=median_steps_work_or_weekend)
plt.xlabel('median_steps_by_month_encoded')
plt.ylabel('median_steps_work_or_weekend')
plt.show()
```
